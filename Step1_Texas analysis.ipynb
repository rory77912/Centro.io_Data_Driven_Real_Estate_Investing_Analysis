```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('tableau-colorblind10')
```

```python
# set graphics parameter
large=22; med =16; small = 12

params= {'figure.figsize':(16,10),
        'axes.titlesize':large,
        'axes.labelsize':med,
        'legend.fontsize':med,
        'xtick.labelsize':small,
        'ytick.labelsize':small}

plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style('white')
```

```python
t = pd.read_csv('Texas_all_data.csv')
```

```python
t1=t[['Num_units','Parking_per_unit','Price_per_sqft','10_yr_treasury_rate']]
t1.columns = ['Num_units', 'Parking_per_unit', 'Price_per_sqft','tenyr_treasury_rate']


#Clean the null data
t1 = t1.dropna(axis=0, subset=['Num_units'])
t1 = t1.dropna(axis=0, subset=['Parking_per_unit'])
t1 = t1.dropna(axis=0, subset=['Price_per_sqft'])
t1 = t1.dropna(axis=0, subset=['tenyr_treasury_rate'])

#Turn values into float
t1['Num_units']= t1['Num_units'].astype(float)
t1['Parking_per_unit']= t1['Parking_per_unit'].astype(float)
t1['Price_per_sqft']= t1['Price_per_sqft'].astype(float)
t1['tenyr_treasury_rate']= t1['tenyr_treasury_rate'].astype(float)
```

```python
t1.columns = ['Num_units', 'Parking_per_unit', 'Price_per_sqft','10_yr_treasury_rate']

plt.figure()
#kind=reg add regression lines
sns.pairplot(t1,
             x_vars = ['Num_units', 'Parking_per_unit','10_yr_treasury_rate'],
             y_vars = ['Num_units', 'Parking_per_unit','10_yr_treasury_rate'],
             hue = 'Price_per_sqft'
            )
#plt.savefig('./test8.jpg')
plt.show()
```
