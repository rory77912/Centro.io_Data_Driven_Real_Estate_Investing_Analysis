```python
data_f = data_f[data_f.Price_per_sqft>=0]
```

```python
data_f = data_f[['Amenities', 'Price_per_sqft']]
```

```python
data_f = data_f.dropna()
```

```python
data_f = data_f[data_f.Price_per_sqft<=216]
```

```python
data_f['fitnesscenter'] = data_f['Amenities'].str.contains(r'Fitness', na=True)
data_f['fitnesscenter'] = data_f['fitnesscenter'].map({True: 1, False: 0})

data_f['laundryfacilities'] = data_f['Amenities'].str.contains(r'Laundry', na=True)
data_f['laundryfacilities'] = data_f['laundryfacilities'].map({True: 1, False: 0})

data_f['propertymanager'] = data_f['Amenities'].str.contains(r'Manager', na=True)
data_f['propertymanager'] = data_f['propertymanager'].map({True: 1, False: 0})
```

```python
data_f['keywords']= data_f['fitnesscenter']+data_f['laundryfacilities']+data_f['propertymanager']
data_f.head()
```
