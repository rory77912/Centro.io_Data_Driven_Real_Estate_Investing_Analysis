```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('tableau-colorblind10')
```

```python
# set graphics parameter
large=22; med =16; small = 12

params= {'figure.figsize':(16,10),
        'axes.titlesize':large,
        'axes.labelsize':med,
        'legend.fontsize':med,
        'xtick.labelsize':small,
        'ytick.labelsize':small}

plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style('white')
```

```python
f = pd.read_csv('Georgia_all_data.csv')
```

```python
f1=f[['Num_units','Parking_per_unit','Price_per_sqft','10_yr_treasury_rate']]
f1.columns = ['Num_units', 'Parking_per_unit', 'Price_per_sqft','tenyr_treasury_rate']


#Clean the null data
f1 = f1.dropna(axis=0, subset=['Num_units'])
f1 = f1.dropna(axis=0, subset=['Parking_per_unit'])
f1 = f1.dropna(axis=0, subset=['Price_per_sqft'])
f1 = f1.dropna(axis=0, subset=['tenyr_treasury_rate'])

#Turn values into float
f1['Num_units']= f1['Num_units'].astype(float)
f1['Parking_per_unit']= f1['Parking_per_unit'].astype(float)
f1['Price_per_sqft']= f1['Price_per_sqft'].astype(float)
f1['tenyr_treasury_rate']= f1['tenyr_treasury_rate'].astype(float)
```

```python
f1.columns = ['Num_units', 'Parking_per_unit', 'Price_per_sqft','10_yr_treasury_rate']

plt.figure()
#kind=reg add regression lines
sns.pairplot(f1,
             x_vars = ['Num_units', 'Parking_per_unit','10_yr_treasury_rate'],
             y_vars = ['Num_units', 'Parking_per_unit','10_yr_treasury_rate'],
             hue = 'Price_per_sqft'
            )
#plt.savefig('./test8.jpg')
plt.show()
```
