```python
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in data_t.Amenities: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
  
wordcloud = WordCloud(background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10,
                max_words = 10).generate(text) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
data_t['fitnesscenter'] = data_t['Amenities'].str.contains(r'Fitness', na=True)
data_t['fitnesscenter'] = data_t['fitnesscenter'].map({True: 1, False: 0})

data_t['businesscenter'] = data_t['Amenities'].str.contains(r'Business', na=True)
data_t['businesscenter'] = data_t['businesscenter'].map({True: 1, False: 0})

data_t['laundryfacilities'] = data_t['Amenities'].str.contains(r'Laundry', na=True)
data_t['laundryfacilities'] = data_t['laundryfacilities'].map({True: 1, False: 0})

data_t['propertymanager'] = data_t['Amenities'].str.contains(r'Manager', na=True)
data_t['propertymanager'] = data_t['propertymanager'].map({True: 1, False: 0})
```

```python
x = data_t['fitnesscenter']
y = data_t['Price_per_sqft']
plt.scatter(x, y)

plt.xticks(range(0,2,1))

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

ax = plt.gca()
ax.set_title("Fitness Center's Impact on Price per Sqft in Texas")
plt.show()
```
